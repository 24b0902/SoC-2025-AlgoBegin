Next lines: student_name subject1:grade1,subject2:grade2,...


**Output Format:**

Line 2: List of students sorted by average grade (descending)

2. Analyze a text to find word frequencies, unique words, and sentence statistics.

**Input Format:**

Multiple lines of text (terminated by empty line or EOF)


**Output Format:**

Line 1: Dictionary of word frequencies (sorted by word)

Line 2: Set of unique words with length > 4

Line 3: List of sentence lengths (number of words per sentence)


**Example:**

**Input:**

The quick brown fox jumps over the lazy dog.

This is a simple sentence.

**Output:**

{'The': 1, 'This': 1, 'a': 1, 'brown': 1, 'dog': 1, 'fox': 1, 'is': 1, 'jumps': 1, 'lazy': 1, 'over': 1, 'quick': 1, 'sentence': 1, 'simple': 1, 'the': 1}

{'brown', 'jumps', 'quick', 'sentence', 'simple'}

[9, 5]

```python

import string

lines = []
while True:
  try:
    line = input()
    if not line:
      break
    lines.append(line)
  except EOFError:
    break

text = " ".join(lines)
text = text.translate(str.maketrans('', '', string.punctuation))
words = text.split()

word_freq = {}
for word in words:
  word_freq[word] = word_freq.get(word, 0) + 1

sorted_word_freq = dict(sorted(word_freq.items()))
print(sorted_word_freq)

unique_long_words = set()
for word in words:
  if len(word) > 4:
    unique_long_words.add(word)
print(unique_long_words)

sentences = " ".join(lines).split('.')
sentence_lengths = []
for sentence in sentences:
  if sentence.strip():
    sentence_words = sentence.strip().translate(str.maketrans('', '', string.punctuation)).split()
    sentence_lengths.append(len(sentence_words))

sentence_lengths
```

3. Analyze a social network to find mutual friends, popular users, and connection statistics.

**Input Format:**

First line: number of friendships

Next lines: user1 user2 (friendship pairs)

Last line: two usernames to check mutual friends


**Output Format:**

Line 1: Dictionary of each user's friends (as sets)

Line 2: List of users sorted by number of friends (descending)

Line 3: Set of mutual friends between the two specified users

Line 4: Tuple of (total_users, total_friendships, average_friends_per_user)


**Example:**

**Input:**

5

Alice Bob

Bob Charlie

Alice Charlie

Charlie David

Bob David

Alice Charlie


**Output:**

{'Alice': {'Bob', 'Charlie'}, 'Bob': {'Alice', 'Charlie', 'David'}, 'Charlie': {'Alice', 'Bob', 'David'}, 'David': {'Bob', 'Charlie'}}

['Bob', 'Charlie', 'Alice', 'David']

{'Bob'}

(4, 5, 2.5)

```python
no_of_friendships = int(input())
social_network = {}

for _ in range(no_of_friendships):
    user1, user2 = input().split()
    if user1 not in social_network:
        social_network[user1] = set()
    if user2 not in social_network:
        social_network[user2] = set()
    social_network[user1].add(user2)
    social_network[user2].add(user1)

print(social_network)

popular_users = sorted(social_network, key=lambda x: len(social_network[x]), reverse=True)
print(popular_users)

user_a, user_b = input().split()
if user_a in social_network and user_b in social_network:
  mutual_friends = social_network[user_a].intersection(social_network[user_b])
  print(mutual_friends)
else:
  print(set())
total_users = len(social_network)
total_friendships = no_of_friendships
average_friends_per_user = sum(len(friends) for friends in social_network.values()) / total_users if total_users > 0 else 0

print((total_users, total_friendships, average_friends_per_user))
```

4. Manage a store inventory with operations like adding stock, processing orders, and generating reports.

**Input Format:**

First line: number of initial inventory items

Next lines: item_name:quantity:price

Following line: number of operations

Next lines: operation_type:parameters


Operations:

ADD:item_name:quantity:price

SELL:item_name:quantity

RESTOCK:item_name:quantity

CHECK:item_name


**Output Format:**

Line 1: Final inventory dictionary

Line 2: List of items that need restocking (quantity < 5)

Line 3: Set of items that were sold

Line 4: Tuple of (total_items, total_value, most_expensive_item)


**Example:**

**Input:**

3

Laptop:10:1000

Mouse:25:20

Keyboard:15:50

4

SELL:Laptop:2

ADD:Monitor:5:300

RESTOCK:Mouse:10

SELL:Keyboard:5

**Output:**

{'Laptop': (8, 1000), 'Mouse': (35, 20), 'Keyboard': (10, 50), 'Monitor': (5, 300)}

['Monitor']

{'Laptop', 'Keyboard'}

(4, 10800, 'Laptop')

```python
inventory = {}
sold_items = set()

num_initial_items = int(input())
for _ in range(num_initial_items):
    item_info = input().split(':')
    item_name = item_info[0]
    quantity = int(item_info[1])
    price = int(item_info[2])
    inventory[item_name] = (quantity, price)

num_operations = int(input())
for _ in range(num_operations):
    operation = input().split(':')
    op_type = operation[0]

    if op_type == 'ADD':
        item_name = operation[1]
        quantity = int(operation[2])
        price = int(operation[3])
        inventory[item_name] = (quantity, price)
    elif op_type == 'SELL':
        item_name = operation[1]
        quantity_to_sell = int(operation[2])
        if item_name in inventory:
            current_quantity, price = inventory[item_name]
            new_quantity = current_quantity - quantity_to_sell
            inventory[item_name] = (new_quantity, price)
            sold_items.add(item_name)
    elif op_type == 'RESTOCK':
        item_name = operation[1]
        quantity_to_restock = int(operation[2])
        if item_name in inventory:
            current_quantity, price = inventory[item_name]
            new_quantity = current_quantity + quantity_to_restock
            inventory[item_name] = (new_quantity, price)
    elif op_type == 'CHECK':
        item_name = operation[1]
        if item_name in inventory:
            print(f"{item_name}: {inventory[item_name][0]} quantity at ${inventory[item_name][1]}")
        else:
            print(f"{item_name} not found in inventory.")

print(inventory)

needs_restocking = [item for item, (quantity, _) in inventory.items() if quantity < 5]
print(needs_restocking)

print(sold_items)

total_items = sum(quantity for quantity, _ in inventory.values())
total_value = sum(quantity * price for quantity, price in inventory.values())
most_expensive_item = max(inventory, key=lambda item: inventory[item][1]) if inventory else None

print((total_items, total_value, most_expensive_item))
```

5. Create a course scheduling system that handles prerequisites, time conflicts, and student enrollments.

**Input Format:**

First line: number of courses

Next lines: course_code:course_name:credits:time_slot:max_students:prerequisites

Following line: number of students

Next lines: student_id:preferred_courses (comma-separated)


**Output Format:**

Line 1: Dictionary of course information

Line 2: List of courses sorted by enrollment demand

Line 3: Set of courses with prerequisite conflicts

Line 4: Tuple of (total_enrollments, courses_at_capacity, average_credits_per_student)


**Example:**

**Input:**

4

CS101:Intro Programming:3:MWF9:30:

CS201:Data Structures:3:MWF10:25:CS101

MATH101:Calculus:4:TTH9:40:

PHYS101:Physics:4:TTH10:35:MATH101

3

S001:CS101,MATH101

S002:CS201,PHYS101

S003:CS101,CS201,MATH101


**Output:**

{'CS101': {'name': 'Intro Programming', 'credits': 3, 'time': 'MWF9', 'max': 30, 'prereq': set(), 'enrolled': {'S001', 'S003'}}, 'CS201': {'name':
'Data Structures', 'credits': 3, 'time': 'MWF10', 'max': 25, 'prereq': {'CS101'}, 'enrolled': {'S003'}}, 'MATH101': {'name': 'Calculus', 'credits': 4,
'time': 'TTH9', 'max': 40, 'prereq': set(), 'enrolled': {'S001', 'S003'}}, 'PHYS101': {'name': 'Physics', 'credits': 4, 'time': 'TTH10', 'max': 35,
'prereq': {'MATH101'}, 'enrolled': set()}}

['CS101', 'MATH101', 'CS201', 'PHYS101']

{'CS201', 'PHYS101'}

(5, 0, 6.67)

```python
num_courses = int(input())
courses = {}
for _ in range(num_courses):
  course_info = input().split(':')
  code = course_info[0]
  name = course_info[1]
  credits = int(course_info[2])
  time_slot = course_info[3]
  max_students = int(course_info[4])
  prerequisites = set(course_info[5].split(',')) if course_info[5] else set()
  courses[code] = {
      'name': name,
      'credits': credits,
      'time': time_slot,
      'max': max_students,
      'prereq': prerequisites,
      'enrolled': set()
  }

num_students = int(input())
student_preferences = {}
for _ in range(num_students):
    student_info = input().split(':')
    student_id = student_info[0]
    preferred_courses = student_info[1].split(',') if student_info[1] else []
    student_preferences[student_id] = preferred_courses

for student_id, preferred_courses in student_preferences.items():
    enrolled_courses_for_student = set()
    for course_code in preferred_courses:
        if course_code in courses:
            course = courses[course_code]
            time_conflict = False
            for enrolled_course_code in enrolled_courses_for_student:
                if courses[enrolled_course_code]['time'] == course['time']:
                    time_conflict = True
                    break

            if not time_conflict and len(course['enrolled']) < course['max']:
                has_prereqs = True
                for prereq in course['prereq']:
                    if prereq not in [c for c in enrolled_courses_for_student]:
                         has_prereqs = False
                         break
                if has_prereqs:
                  course['enrolled'].add(student_id)
                  enrolled_courses_for_student.add(course_code)


print(courses)

courses_by_demand = sorted(courses, key=lambda code: len(courses[code]['enrolled']), reverse=True)
print(courses_by_demand)

prereq_conflicts = set()
for student_id, preferred_courses in student_preferences.items():
    enrolled_courses_for_student = set()
    for course_code in preferred_courses:
        if course_code in courses:
            course = courses[course_code]
            time_conflict = False
            for enrolled_course_code in enrolled_courses_for_student:
                if courses[enrolled_course_code]['time'] == course['time']:
                    time_conflict = True
                    break
            if time_conflict:
              continue
            prereq_met = True
            for prereq in course['prereq']:
              if prereq not in preferred_courses:
                prereq_met = False
                break

            if not prereq_met:
                prereq_conflicts.add(course_code)
            elif len(course['enrolled']) < course['max']:
                 course['enrolled'].add(student_id)
                 enrolled_courses_for_student.add(course_code)


print(prereq_conflicts)

total_enrollments = sum(len(course['enrolled']) for course in courses.values())
courses_at_capacity = sum(1 for course in courses.values() if len(course['enrolled']) == course['max'])
total_credits = sum(courses[course_code]['credits'] for course_code in courses for student_id in courses[course_code]['enrolled'])
total_students_with_enrollment = len(student_preferences)
average_credits_per_student = round(total_credits / total_students_with_enrollment, 2) if total_students_with_enrollment > 0 else 0

print((total_enrollments, courses_at_capacity, average_credits_per_student))
```

7. Manage a tournament bracket system with player rankings, match results, and statistics.

**Input Format:**

First line: number of players

Next lines: player_name:initial_rating

Following line: number of matches played

Next lines: player1:player2:winner:score1:score2


**Output Format:**

Line 1: Dictionary of player statistics (wins, losses, rating)

Line 2: List of players ranked by final rating

Line 3: Set of players who beat higher-rated opponents

Line 4: Tuple of matches played by each player (as tuple pairs)

Line 5: Tournament statistics (total_matches, upsets, average_rating_change)


**Example:**

**Input:**

4

Alice:1200

Bob:1100

Charlie:1300

David:1000

3

Alice:Bob:Alice:3:1

Charlie:David:Charlie:2:0

Alice:Charlie:Charlie:1:2


**Output:**

{'Alice': {'wins': 1, 'losses': 1, 'rating': 1190}, 'Bob': {'wins': 0, 'losses': 1, 'rating': 1090}, 'Charlie': {'wins': 2, 'losses': 0, 'rating':
1320}, 'David': {'wins': 0, 'losses': 1, 'rating': 990}}

['Charlie', 'Alice', 'Bob', 'David']

set()

[('Alice', 2), ('Bob', 1), ('Charlie', 2), ('David', 1)]

(3, 0, 15.0)

```python

player_stats = {}
matches_played = {}
upsets = set()

num_players = int(input())
for _ in range(num_players):
    player_info = input().split(':')
    player_name = player_info[0]
    initial_rating = int(player_info[1])
    player_stats[player_name] = {'wins': 0, 'losses': 0, 'rating': initial_rating}
    matches_played[player_name] = 0

num_matches = int(input())
total_rating_change = 0
for _ in range(num_matches):
    match_info = input().split(':')
    player1_name = match_info[0]
    player2_name = match_info[1]
    winner_name = match_info[2]
    score1 = int(match_info[3])
    score2 = int(match_info[4])

    matches_played[player1_name] += 1
    matches_played[player2_name] += 1

    player1_rating = player_stats[player1_name]['rating']
    player2_rating = player_stats[player2_name]['rating']
    k = 30
    expected_score1 = 1 / (1 + 10**((player2_rating - player1_rating) / 400))
    expected_score2 = 1 / (1 + 10**((player1_rating - player2_rating) / 400))

    if winner_name == player1_name:
        player_stats[player1_name]['wins'] += 1
        player_stats[player2_name]['losses'] += 1
        rating_change = k * (1 - expected_score1)
        if player1_rating < player2_rating:
            upsets.add(player1_name)
    else:
        player_stats[player1_name]['losses'] += 1
        player_stats[player2_name]['wins'] += 1
        rating_change = k * (0 - expected_score1)
        if player2_rating < player1_rating:
            upsets.add(player2_name)

    player_stats[player1_name]['rating'] += rating_change
    player_stats[player2_name]['rating'] -= rating_change
    total_rating_change += abs(rating_change) * 2

for player in player_stats:
    player_stats[player]['rating'] = int(player_stats[player]['rating'])

print(player_stats)

ranked_players = sorted(player_stats, key=lambda player: player_stats[player]['rating'], reverse=True)
print(ranked_players)

print(upsets)

matches_tuple = [(player, matches_played[player]) for player in matches_played]
print(matches_tuple)

total_matches = num_matches
num_upsets = len(upsets)
average_rating_change = round(total_rating_change / (2 * num_matches) if num_matches > 0 else 0, 2)

print((total_matches, num_upsets, average_rating_change))
```
