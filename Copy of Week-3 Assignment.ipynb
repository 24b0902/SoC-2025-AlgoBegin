**1.** Linear Search in a List

Write a function to perform linear search on a list of integers.

**Input:** A list of n integers and a target integer x

**Output:** Index of x if found, else -1

```python

def linearSearch():
  inputList=(input("Enter a list of integers separated by spaces: ").split())
  target=(input("Enter the target integer: "))
  for i in range(len(inputList)):
    if inputList[i]==target:
      return i
  return -1
ans=linearSearch()
print(ans)
```

**2.** Recursive Factorial

Write a recursive function to calculate the factorial of a number.

**Input:** Integer n

**Output:** n!

```python
def factorial(n):
  if n==0:
    return 1
  else:
    return n*factorial(n-1)
n=int(input("Enter a number: "))
print(factorial(n))
```

**3.** Binary Search (Recursive Implementation)

Write a recursive binary search function for a sorted list.

**Input:** Sorted list of integers and a target value

**Output:** Index of the target or -1 if not found

```python
def binarySearch(arr,low,high,target):
  mid=round((low+high)/2)
  while(low<=high):
    if target==arr[mid]:
      return mid
    elif target<arr[mid]:
      return binarySearch(arr,low,mid-1,target)
    else:
      return binarySearch(arr,mid+1,high,target)
  return -1
arr=list(map(int,input("Enter a sorted list of integers separated by spaces: ").split()))
target=int(input("Enter the target integer: "))
print(binarySearch(arr,0,len(arr)-1,target))
```

**4.** Insertion Sort Implementation

Implement the insertion sort algorithm without using built-in sort.

**Input:** Unsorted list of integers

**Output:** Sorted list

```python
def insertionSort(arr):
  if(len(arr)>1):
    for i in range(1,len(arr)):
      j=i-1
      key=arr[i]
      while(j>=0 and key<arr[j]):
        arr[j+1]=arr[j]
        j-=1
      arr[j+1]=key
  return arr
arr=list(map(int,input("Enter a list of integers separated by spaces: ").split()))
print(insertionSort(arr))
```

**5.** Merge Sort using Recursion

Implement merge sort using recursion.

**Input:** A list of integers

**Output:** Sorted list

```python
def mergeSort(arr,start,end):
  if(start<end):
    mid=(start+end)//2
    mergeSort(arr,start,mid)
    mergeSort(arr,mid+1,end)
    merge(arr,start,mid,end)
  return arr
def merge(arr,start,mid,end):
  temp=[]
  i=start
  j=mid+1
  while i<=mid and j<=end:
    if arr[i]<=arr[j]:
      temp.append(arr[i])
      i+=1
    else:
      temp.append(arr[j])
      j+=1
  while i<=mid:
    temp.append(arr[i])
    i+=1
  while j<=end:
    temp.append(arr[j])
    j+=1
  for i in range(start,end+1):
    arr[i]=temp[i-start]

arr=list(map(int,input("Enter a list of integers separated by spaces: ").split()))
print(mergeSort(arr,0,len(arr)-1))


```

**6.** Count Number of Subsets (Backtracking)

Given a list of positive integers and a target sum, count the number of subsets whose sum is exactly equal to the target.

```python
arr=list(map(int,input("Enter a list of integers separated by spaces: ").split()))
target=int(input("Enter the target integer: "))
def countSubsets(arr, target, index=0, current_sum=0):
    if current_sum == target:
        return 1
    if index >= len(arr) or current_sum > target:
        return 0
    include = countSubsets(arr, target, index + 1, current_sum + arr[index])
    exclude = countSubsets(arr, target, index + 1, current_sum)

    return include + exclude
print("Number of subsets:", countSubsets(arr, target))
```
