**2.**Given a string, print the number of vowels in it.

**Input Format:**

A single string (lowercase only).

**Output Format:**

A single integer â€” count of vowels.

```python
input_string=input("Enter a string in lowercase only:")
count=0
for char in input_string:
  if char=="a" or char=="e" or char=="i" or char=="o" or char=="u":
    count+=1
print(count)
```
